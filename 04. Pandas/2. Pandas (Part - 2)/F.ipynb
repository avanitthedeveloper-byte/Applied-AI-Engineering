```python
# 6. Melt & Pivot for Reshaping
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "country": ["USA", "USA", "India", "India"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100, 120, 90, 110],
    "profit": [20, 25, 10, 22],
    # "tax": [10, 20, 30, 40]
})

df
```

```python
melted_df = df.melt(
    id_vars = ["country", "year"],
    value_vars = ["sales", "profit"],
    var_name = "metrics",
    value_name = "value",
)

melted_df
```

```python
melted_df.pivot(
    index = ["country", "year"],
    columns = "metrics",
    values = "value"
)
```
