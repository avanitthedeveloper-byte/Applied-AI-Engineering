```python
# 9. Data Concatenation with Pandas
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "id" : [1, 2, 3],
    "name" : ["Adom", "Eve", "Bob"]
})

df2 = pd.DataFrame({
    "id" : [4, 5, 6],
    "name" : ["Charlie", "Fester", "Gilian"]
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2]) # Row
```

```python
pd.concat([df1, df2], ignore_index= True)
```

```python
pd.concat([df1, df2], axis=1) # Column
```
