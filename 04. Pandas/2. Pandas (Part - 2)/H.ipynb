```python
import pandas as pd
```

```python
df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Adom", "Bob", "Charlie", "Dave"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [2, 1, 4, 5],
    "amount": [250, 120, 103, 180]
})

pd.merge(df_customers, df_orders, on="customer_id") # Inner Join
```

```python
pd.merge(df_customers, df_orders, on="customer_id", how="left") # Left Join
```

```python
pd.merge(df_customers, df_orders, on="customer_id", how="right") # Righrt Join
```

```python
pd.merge(df_customers, df_orders, on="customer_id", how="outer") # Inner Join
```

```python
pd.merge(df_customers, df_orders)
```
