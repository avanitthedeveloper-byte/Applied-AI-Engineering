```python
# 5. Group by & Aggregation
```

```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")

df
```

```python
df2 = df.copy()

df2
```

```python
# Group by

df2.groupby("country")["income"].mean()
```

```python
df2.groupby("country")["income"].min()
```

```python
df2.groupby("country")["income"].max()
```

```python
df2.groupby("gender")["income"].mean()
```

```python
# Aggregation
df2.groupby("country")["income"].agg(["mean", "min", "max"])
```

```python
df2.groupby("country")["income"].agg(["mean", "min", "max"])
```

```python
# Note - Label can also be changed

df2.groupby("country")["income"].agg(Mean_Value="mean", Min_Value="min", Max_Value="max")
```

```python
df2.groupby("country").agg({
    "income" : "mean",
    "age" : "mean"
})
```

```python
df2.groupby("country").agg({
    "income" : "max",
    "age" : "mean"
})
```

```python
# Note - Add Label

df2.groupby("country").agg(
    Max_Salary = ("income", "max"),
    Mean_Age = ("age", "mean")
)
```
