```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")

df
```

```python
df2 = df.copy()

# rename (columns) All Values of columns
df2.columns = ["Id", "Name", "Age", "Country", "Gender", "Income"]

# rename (columns) Specific values of columns
df2 = df2.rename(columns={"Income": "Salary"})

# rename (row) Specific values of columns
# df2 = df2.rename(index={1: "One"})

# sort_values
df2 = df2.sort_values("Salary")

# df2 = df2.sort_values(["Salary", "Age"])

# sort_index
df2 = df2.sort_index()

# reset_index
df2 = df2.reset_index()

# rank()
df2["Salary"].rank()

# Reorder Columns
df2[["Id", "Name", "Age", "Country", "Salary", "Gender"]]
```
