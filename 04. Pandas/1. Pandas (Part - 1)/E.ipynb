```python
import pandas as pd
```

```python
cgpa = pd.Series([9.2, 9.7, 9.3, 9.9], index=["Cipher Mind", "Mind", "Cipher", "Mind Cipher"])

print(cgpa, type(cgpa))
```

```python
# 1. Homogenous
cgpa = pd.Series([9.2, 9.7, 9.3, 9.9, 9.0], index=["Cipher Mind", "Mind", "Cipher", "Mind Cipher", 3])

print(cgpa, type(cgpa))
```

```python
# 2. Vectorized Operation
print(cgpa + 0.1)
```

```python
# 4. Mutable Values, Immutable Size
cgpa = pd.Series([9.2, 9.7, 9.3, 9.9, 9.0], index=["Cipher Mind", "Mind", "Cipher", "Mind Cipher", 3])
print(cgpa, type(cgpa))
cgpa[3] = 6.5
print(cgpa, type(cgpa))
```

```python
# Note - Drop Function
cgpa.drop(3)
```
