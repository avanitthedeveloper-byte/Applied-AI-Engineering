```python
# 12. Multi-dimensional Arrays & Axes
```

```python
import numpy as np
```

```python
arr1D = np.array([1,3,5])

print(arr1D)
```

```python
arr2D = np.array([[1,3,5], [2,7,1], [1,5,7]])

print(arr2D)

print(f"Sum of all Elements: {np.sum(arr2D)}")
print(f"Sum of all vertical Elements: {np.sum(arr2D, axis=0)}")
print(f"Sum of all horizontal Elements: {np.sum(arr2D, axis=1)}")
```

```python
# Slicing in 2D Arrays
print(arr2D[0:2, 1:2])
```
