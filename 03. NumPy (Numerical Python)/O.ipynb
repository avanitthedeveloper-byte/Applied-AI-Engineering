```python
import numpy as np
```

```python
# Normalization

arr = np.array([[1, 2], [3, 4]])

mean = np.mean(arr)
print(f"Mean: {mean}")
std_dev = np.std(arr)
print(f"Standard Deviation: {std_dev}")

normalized_arr = (arr-mean)/std_dev

print(normalized_arr)
```
