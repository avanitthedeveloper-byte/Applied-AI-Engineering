```python
import numpy as np
```

```python
list = [1,2,3,4,5,6]

sub_list = list[1:4]

sub_list[1] = 900
print(sub_list)
print(list)
```

```python
new_arr = np.array(list)

# sub_list = new_arr[1:4]
sub_list = new_arr[1:4].copy()

sub_list[1] = 900
print(sub_list)
print(new_arr)
```
