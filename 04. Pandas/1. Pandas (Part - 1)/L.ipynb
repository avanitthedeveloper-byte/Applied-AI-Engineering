```python
import pandas as pd
```

```python
df = pd.read_csv("globalAirQuality.csv")

df
```

```python
df.query("aqi > 100")
```

```python
df.query("aqi > 100 & temperature < 30")[["city", "country"]]
```

```python
aqi_val = 100
copy_df=df.query("aqi > @aqi_val & temperature < 30")[["city", "country"]].copy()

copy_df
```
