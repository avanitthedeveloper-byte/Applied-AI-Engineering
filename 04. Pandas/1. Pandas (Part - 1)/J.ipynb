```python
import pandas as pd
```

```python
df = pd.read_csv("globalAirQuality.csv")

df
```

```python
# 1. Select Columns
df["country"]
```

```python
df[["country", "city"]]
```

```python
# 2. Select Rows (Label & Index based) - loc & iloc
df.loc[3]
```

```python
df.loc[3:18] # starting idx : Ending idx ("Inclusive")
```

```python
df.iloc[3]
```

```python
df.iloc[3:18] # starting idx : Ending idx ("Non-inclusive")
```

```python
# 3. Select Rows & Columns
df.loc[1500, ["country", "city"]]

df.iloc[1500, [1, 2]] # Pass the index of country and city of data
```

```python
df.loc[0:25, ["country", "city", "latitude", "longitude"]]
```

```python
df.iloc[0:25, [1, 2, 3, 4]]
```

```python
df.iloc[0:25, 1:5]
```

```python
# 4. Select single scalar value - at & iat
df.at[0, "city"]
```

```python
df.iat[0, 2]
```
