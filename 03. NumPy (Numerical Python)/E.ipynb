```python
# 5. Creating Arrays (using built-in methods)
```

```python
import numpy as np
```

```python
arr = np.zeros((2, 3))

print(arr, arr.shape, arr.dtype)
```

```python
arr = np.ones((4, 5))

print(arr, arr.shape, arr.dtype)
```

```python
arr = np.full((4, 5), "Cipher Mind")

print(arr, arr.shape, arr.dtype)
```

```python
arr = np.ones((4, ))

print(arr, arr.shape, arr.dtype)
```

```python
arr = np.ones(( 5, 1))

print(arr, arr.shape, arr.dtype)
```

```python
arr = np.eye((3))

print(arr, arr.shape, arr.dtype)
```

```python
arr = np.arange(2, 9, 3)

print(arr)
```

```python
arr = np.linspace(1, 50, 5)

print(arr)
```
