```python
# 3. Python List v/s NumPy Arrays
```

```python
import numpy as np
import time
```

```python
# In terms of Time
size = 1_000_000

#python list
py_list = list(range(size))
start = time.time()

py_list = [val**2 for val in py_list]
end = time.time()

print(f"Python List Time: {end-start}")


np_array = np.array(py_list)
start1 = time.time()

np_array = np_array**2
end1 = time.time()

print(f"Numpy Array Time: {end1-start1}")
```

```python
# In terms of Size
import sys

print(f"Size of Python List: {sys.getsizeof(py_list)}")

print(f"Size of Numpy Array: {sys.getsizeof(np_array)}")
```
