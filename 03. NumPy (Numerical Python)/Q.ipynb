```python
import numpy as np
```

```python
 # Aggregation
arr = np.array([1, 3, 5, 7, 3, 9])

print(arr)
print(f"Sum(): {np.sum(arr)}")
print(f"Prod(): {np.prod(arr)}")
print(f"Min(): {np.min(arr)}")
print(f"Max(): {np.max(arr)}")
print(f"Mean(): {np.mean(arr)}")
print(f"Standard Deviation(): {np.std(arr)}")
print(f"Variance(): {np.var(arr)}")
print(f"Median(): {np.median(arr)}")
print(f"Argmin(): {np.argmin(arr)}")
print(f"Argmax(): {np.argmax(arr)}")
```

```python
# Power
print(arr)
print(f"Square(): {np.square(arr)}")
print(f"SQRT(): {np.sqrt(arr)}")
print(f"Pow(): {np.pow(arr, 3)}")
```

```python
# Rounding
print(arr)
print(f"Round(): {np.round(np.sqrt(arr))}")
print(f"Floor(): {np.floor(np.sqrt(arr))}")
print(f"Ceil(): {np.ceil(np.sqrt(arr))}")
print(f"Trunc(): {np.trunc(np.sqrt(arr))}")
```

```python
# Log & Exponential
print(arr)
print(f"Log(): {np.log(arr)}")
print(f"Log10(): {np.log10(arr)}")
print(f"Log2(): {np.log2(arr)}")
print(f"Exp(): {np.exp(arr)}")
```
