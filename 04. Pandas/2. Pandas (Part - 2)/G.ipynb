```python
# 7. Basic Visualization with Pandas
```

```python
import pandas as pd
```

```python
df = pd.read_csv("employee_data.csv")
df
```

```python
df["Age"].hist()
```

```python
df.plot(x = "Age", y = "Salary")
```

```python
df.plot(kind="scatter", x = "Age", y = "Salary")
```

```python
df.plot(kind="kde", x = "Age", y = "Salary")
```

```python
df.plot(kind="pie", x = "Age", y = "Salary")
```

```python
df.plot(kind="hexbin", x = "Age", y = "Salary")
```
