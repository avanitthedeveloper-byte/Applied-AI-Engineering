```python
import json
```

```python
# load the data
def load_data(file_name):
    with open(file_name, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
print(data)
```

```python
# clean & structure the data
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    cleaned_data = []
    unique_users = set()
    for user in data:
        # clean ratings
        row_rating = str(user["rating"]).strip().lower()
        if row_rating in text_to_num:
            row_rating = text_to_num[row_rating]
        user["rating"] =  row_rating

        # handle missing values
        row_age = user.get("age")
        if row_age == None:
            user["age"] = None

        #Deduplication
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
```

```python
clean_data(data)
```
