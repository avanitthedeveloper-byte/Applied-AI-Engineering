```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")

df
```

```python
df2 = df.copy()
# apply()
df2["tax"] = df2["income"].apply(lambda x: "20%" if x >= 60000 else "10%")

# map()
gender_map = {"Male" : "M", "Female" : "F", "Unknown" : "U", "NaN" : "M"}

df2["gender"] = df2["gender"].map(gender_map)

# assign()
df2 = df2.assign(paying_tax = "yes")

# replace(old, new)

df2["tax"] = df2["tax"].replace("10%", "12%")

df2
```
