```python
import pandas as pd
```

```python
info = {
    "Name": ["Cipher", "Mind", "Cipher Mind"],
    "Marks": [98, 96, 92],
    "CGPA": [9.8, 9.6, 9.2]
}

print(info)
```

```python
# Method - 1 Dictionary of List

df = pd.DataFrame(info)

print(type(df))

df
```

```python
print(df.index)
```

```python
print(df.columns)
```

```python
# Method - 2 DataFrame as List of Lists

df = pd.DataFrame([["Cipher", 26], ["Mind", 16], ["Cipher Mind", 28]], columns = ["Name", "Age"])

df
```

```python
# Method - 3 numpy array

import numpy as np

numpy_array = np.array([["Cipher", 26], ["Mind", 16], ["Cipher Mind", 28]])

numpy_array

df = pd.DataFrame(numpy_array, columns = ["Name", "Age"])

df
```
